```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

detector tutrial: https://www.bogotobogo.com/python/OpenCV_Python/python_opencv3_Image_Object_Detection_Face_Detection_Haar_Cascade_Classifiers.php

```python
face_cascade = cv2.CascadeClassifier('models/haarcascade_fullbody.xml')
img = cv2.imread('data_samples/sample_img2.jpg', cv2.IMREAD_COLOR)
assert type(img)!=type(None), ('invalid image')

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.005, 5)
print(faces)

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)

# Display the output
plt.imshow(cv2.cvtColor(img[:,:,:], cv2. COLOR_BGR2RGB))
#plt.savefig('assets/my_photo.png', dpi=600)
```
