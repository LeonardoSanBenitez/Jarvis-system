```python
path = 'WESAD/'
subjects = ['S3', 'S4', 'S5']
variables=['ACC', 'BVP', 'EDA', 'HR', 'TEMP']
skip = 10
n=100000+skip
#TODO: for each subject
for subject in subjects:
    df = pd.DataFrame()
    for variable in variables:
        df2 = pd.read_csv(path+subject+'/'+subject+'_E4_Data/'+variable+'.csv', names=[variable], usecols=[0])
        df = pd.concat([df, df2[skip:n]], axis=1)
    df.to_csv(subject+".csv")
    print("done for subject", subject, ", exporting data", df.shape)
```
